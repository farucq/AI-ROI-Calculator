```python
import pandas as pd
import numpy as np

print("Environment Ready")
```

Manual Process Cost Model

```python
employees = 6
hours_per_day = 6
cost_per_hour = 350
days_per_month = 22

monthly_manual_cost = employees * hours_per_day * cost_per_hour * days_per_month
annual_manual_cost = monthly_manual_cost * 12

monthly_manual_cost, annual_manual_cost
```

AI System Cost Model

```python
one_time_cost = 400000
monthly_ai_cost = 45000 + 15000

annual_ai_cost = monthly_ai_cost * 12 + one_time_cost

monthly_ai_cost, annual_ai_cost
```

ROI Formula Engine

```python
roi = ((annual_manual_cost - annual_ai_cost) / annual_ai_cost) * 100
savings = annual_manual_cost - annual_ai_cost
payback_months = one_time_cost / (monthly_manual_cost - monthly_ai_cost)

roi, savings, payback_months
```

Scenario Simulation (Best / Expected / Worst)

```python
scenarios = {
    "Best": {"ai_monthly": 50000, "efficiency_gain": 0.9},
    "Expected": {"ai_monthly": 60000, "efficiency_gain": 0.75},
    "Worst": {"ai_monthly": 75000, "efficiency_gain": 0.6},
}

results = []

for name, s in scenarios.items():
    ai_cost = s["ai_monthly"] * 12 + one_time_cost
    effective_manual = annual_manual_cost * s["efficiency_gain"]
    roi_val = ((effective_manual - ai_cost) / ai_cost) * 100

    results.append([name, round(ai_cost), round(roi_val,2)])

df = pd.DataFrame(results, columns=["Scenario", "Annual AI Cost", "ROI %"])
df
```

Generate Excel ROI Calculator

```python
excel_data = {
    "Manual Annual Cost": [annual_manual_cost],
    "AI Annual Cost": [annual_ai_cost],
    "Savings": [savings],
    "ROI %": [roi],
    "Payback (months)": [payback_months]
}

excel_df = pd.DataFrame(excel_data)
excel_df.to_excel("AI_ROI_Calculator.xlsx", index=False)

print("Excel File Created")
```

```python
summary = f"""
AI ROI ANALYSIS – DOCUMENT PROCESSING

Manual annual cost: ₹{annual_manual_cost:,.0f}
AI system annual cost: ₹{annual_ai_cost:,.0f}

Annual savings: ₹{savings:,.0f}
ROI: {roi:.2f}%
Payback period: {payback_months:.1f} months

Conclusion:
AI automation is financially viable with strong cost recovery within first year.
"""

with open("business_report.txt", "w") as f:
    f.write(summary)

print(summary)
```
